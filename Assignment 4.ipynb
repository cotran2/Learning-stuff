```python
_LOL.head()
```

```python
_LOL_NA = _LOL[_LOL['League']=="North_America"]
```

```python
df = _LOL_NA.groupby(['Season','blueTeamTag'])
df.head()
```

```python
_LOL_NA_1=_LOL_NA[['redTeamTag','rResult','Season']].groupby(['Season','redTeamTag',]).agg({'rResult':['mean','count'],})
```

```python
_LOL_NA_2=_LOL_NA[['blueTeamTag','bResult','Season']].groupby(['Season','blueTeamTag']).agg({'bResult':['mean','count'],})
```

```python
_LOL_NA_2.head(20)
```

```python
_LOL_NA_1=_LOL_NA_1.reset_index()
_LOL_NA_2=_LOL_NA_2.reset_index()
```

```python
df= pd.merge(_LOL_NA_1,_LOL_NA_2,how='inner',left_on=['Season','redTeamTag'],right_on=['Season','blueTeamTag'])
```

```python
df.redTeamTag.unique()
```

```python
df['Total Count'] = df['rResult']['count']+df['bResult']['count']
```

```python
df['Win rate'] = df['rResult']['count']*df['rResult']['mean']+df['bResult']['count']*df['bResult']['mean']
```

```python
df['Win rate']=df['Win rate']/df['Total Count']
```

```python
i=0
sns.set_style("whitegrid")
plt.figure(5,figsize=(18,6))
for season in df['Season'].unique():
    data = df[df['Season']==season]
    i+=1
    plt.subplot(330+i)
    sns.barplot(x='redTeamTag', y="Win rate",data=data)
    plt.axhline(data['Win rate'].mean(), color='b', linestyle='dashed', linewidth=2) #Mean line
    plt.ylabel('Win rate')
    plt.xlabel('Team name')
    plt.title('Win rate by each team during {}'.format(season))
plt.subplots_adjust(top=1.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.show()
    
```
